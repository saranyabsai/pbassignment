```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\StudentsPerformance.csv")
print(data.head(32))
```

```python


# Counting the number of male and female students
race_ethnicity_count = data["race/ethnicity"].value_counts()

# Creating a bar chart
plt.bar(race_ethnicity_count.index, race_ethnicity_count.values)

# Adding title and labels to the chart
plt.title("Race/Ethnicity Distribution")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Number of Students")

# Displaying the chart
plt.show()

```

```python
sns.scatterplot(x="reading score", y="writing score", data=data)
plt.title("Reading Scores vs. Writing Scores")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

```python
data=data['parental level of education']
data.value_counts().plot(kind='pie')
plt.axis('equal')
plt.title('Pie chart of various groups')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\StudentsPerformance.csv")
sns.boxplot(x='gender', y='math score', data=data)
plt.title('Math Scores by Gender')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.hist(data['parental level of education'],edgecolor='black', bins=6)
xt1=("bachelor's degree","some collage","masters degree","associates degree","high school","some high school")
plt.xticks([0.35,1.25,2.05,2.95,3.8,4.6],xt1)
plt.xlabel('Various Degrees',fontsize="10")
plt.ylabel('Count',fontsize="10")
plt.title('Histogram For Parental Level Of Education',fontsize="12")
plt.show()
```
